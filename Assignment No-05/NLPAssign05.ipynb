```python
morphology_rules = [

    {"operation": "plural", "condition": "default", "add": "s", "delete": ""},
    {"operation": "plural", "condition": "endswith_y", "add": "ies", "delete": "y"},


    {"operation": "past", "condition": "default", "add": "ed", "delete": ""},
    {"operation": "ing", "condition": "endswith_e", "add": "ing", "delete": "e"},


    {"operation": "negative", "condition": "default", "add": "un", "delete": "", "prefix": True},
    {"operation": "comparative", "condition": "default", "add": "er", "delete": ""}
]
```

```python
def apply_rule(word, rule):
    base = word

    if rule["delete"] and word.endswith(rule["delete"]):
        base = word[:-len(rule["delete"])]

    if rule.get("prefix", False):
        return rule["add"] + base
    else:
        return base + rule["add"]
```

```python
words = [
    ("cat", "plural"),
    ("carry", "plural"),
    ("play", "past"),
    ("make", "ing"),
    ("happy", "negative"),
    ("big", "comparative")
]
```

```python
print(f"{'Base':<10} {'Operation':<15} {'Add':<8} {'Delete':<8} {'Result'}")
print("-" * 60)

for word, op in words:
    for rule in morphology_rules:
        if rule["operation"] == op:
            if rule["condition"] == "endswith_y" and not word.endswith("y"):
                continue
            if rule["condition"] == "endswith_e" and not word.endswith("e"):
                continue

            result = apply_rule(word, rule)

            print(f"{word:<10} {op:<15} {rule['add']:<8} {rule['delete']:<8} {result}")
            break
```
